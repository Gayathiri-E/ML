```python
# polynomial regression from scratch

import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4,5])
y= np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
deg = 2
```

```python
#design matrix --column stack fn

x_poly = np.column_stack([x**i for i in range(deg+1)])
x_poly
```

```python
weights = np.zeros(deg +1)
weights
```

```python
learning_rate = 0.001
num_iterations = 10000 #epochs
```

```python
for _ in range(num_iterations):
  #calculate prediction
  predictions = np.dot(x_poly,weights)
  print(weights)
  print(predictions)
  #calculate error
  error = predictions - y
  #update weights using gradient descent(used to reduce cost fn and find optimal result)
  gradient = np.dot(x_poly.T , error)/ len(x) #T - transpose
  weights -= learning_rate *gradient
  print(f'weights for iteration {_}: {weights}')
```

```python
def predict(x,weights):
  x_poly = np.array([x**i for i in range(deg + 1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x,y ,label="Data points")
x_range = np.linspace(min(x),max(x),100)
y_pred = [predict(x,weights) for x in x_range]
plt.plot(x_range,y_pred,color ='r',label="Polynomial Regression")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
