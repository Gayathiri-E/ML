```python
data['buying'],_ = pd.factorize(data['buying']) #factorize -- change to numeric value
data['maint'],_ = pd.factorize(data['maint'])
data['doors'],_ = pd.factorize(data['doors'])
data['persons'],_ = pd.factorize(data['persons'])
data['lug_boot'],_ = pd.factorize(data['lug_boot'])
data['safety'],_ = pd.factorize(data['safety'])
data.head()
```

**label encoding**

[a,b,c] -- [0,1,2]--take  variable separately and transform to numeric value

**one hot encoding**

[a,b,c] -- class a,class b, class c -- [0,1,2]---take the variable and make each column and transform to numeric value

```python
x = data.iloc[:,0:-1]
print(x)
y = data.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state =0)#randmly picks the value in train data ---default random state is 42
```

```python
x.columns
```

```python
from sklearn import tree, metrics
dtree = tree.DecisionTreeClassifier(criterion =  'gini') #entrophy r gini
dtree.fit(x_train,y_train)
```

```python
y_train_pred = dtree.predict(x_train)
```

```python
accuracy = metrics.accuracy_score(y_train,y_train_pred) #train accuracy --
print('train accuracy :{:.2f}'.format(accuracy))
```

```python
y_test_pred = dtree.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test,y_test_pred)
print('test accuracy :{:.2f}'.format(accuracy))
```
