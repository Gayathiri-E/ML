```python
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
titanic
```

```python
titanic.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
titanic[titanic.Embarked.isnull()] #embarked has 2 null values
```

```python
titanic.drop(['Cabin'],axis=1,inplace = True)
titanic['Age']=titanic['Age'].fillna(titanic['Age'].median())
titanic['Embarked'] = titanic['Embarked'].fillna('C')
titanic.Sex = titanic.Sex.replace(['male','female'],[0,1])
titanic.Embarked=titanic.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
titanic.dropna(inplace=True) #dropna = drop null values
```

```python
tit_test = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
tit_test
```

```python
tit_test.drop(['Cabin'],axis=1,inplace = True)
tit_test['Age']=tit_test['Age'].fillna(tit_test['Age'].median())
tit_test['Embarked'] = tit_test['Embarked'].fillna('C')
tit_test.Sex = tit_test.Sex.replace(['male','female'],[0,1])
tit_test.Embarked=tit_test.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
tit_test.dropna(inplace=True)
```

```python
features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"] #independent var
```

```python
c = 0.7 #c=gamma
clf = svm.SVC(kernel='linear',C=c).fit(titanic[features],titanic['Survived']) #survived - dependent(y)
```

```python
prediction = clf.predict(tit_test[features])
prediction
```
