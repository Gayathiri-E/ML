```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df[['mean_radius','mean_texture','mean_perimeter','mean_area','mean_smoothness']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
import lightgbm as lgb
clf = lgb.LGBMClassifier(learning_rate=0.01)
clf.fit(x_train,y_train)
```

```python
y_pred =clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accu =accuracy_score(y_pred,y_test)
print(accu)
```

LIGHTGBM REGRESSOR

```python
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


param_grid = {
    'boosting_type': ['gbdt'],
    'num_leaves': [31, 50, 75],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [50, 100, 200],
    'objective': ['regression'],
    'metric': ['l2'],
}


lgb_model = lgb.LGBMRegressor()

grid_search = GridSearchCV(estimator=lgb_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, verbose=1)

grid_search.fit(x_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best negative mean squared error: {:.4f}".format(grid_search.best_score_))

best_lgb_model = grid_search.best_estimator_

y_pred = best_lgb_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on Test Set: {:.4f}".format(mse))

```
