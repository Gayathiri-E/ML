```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('/content/IRIS.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['species'],_=pd.factorize(dataset['species'])
```

```python
dataset
```

```python
X=dataset.iloc[:,[2,3]].values
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elboe Method")
plt.xlabel("No of Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeansmodel=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeansmodel.fit_predict(X)
```

```python
#Visualizing all the clusters

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 150, c = 'yellow', label = 'Centroids')
plt.title('IRIS')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()
plt.show()
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
feature=['petal_length','petal_width']
x=dataset[feature]
```

```python
x_standardized = StandardScaler().fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.5,min_samples = 5)
clusters = dbscan.fit_predict(x_standardized)
clusters
```

```python
centroids = pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x['petal_length'], x['petal_width'], c=clusters, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering of Iris')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```
